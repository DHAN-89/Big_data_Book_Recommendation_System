```
from __future__ import print_function
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```
book = sc.textFile('/FileStore/tables/story-3.txt')
book.count()
```

```
book.collect()
```

```
###Converting all the lines into list of words using the flapmap

split_lines=book.flatMap(lambda e : e.split(" "))
split_lines.count()
```

```
###Converting all words into lower case

lower_case = split_lines.map(lambda x : x.lower() )   
```

```
lower_case.collect()
```

```
# Removing punctuations

remove_punc = lower_case.filter(lambda x :x.replace("[^a-z' ]", '')) 
```

```
Foul_words=['disgusting','revolting','repellent','repulsive','repugnant','abhorrent','loathsome','offensive','detestable','awful','dreadful',
'horrible','terrible','horrendous','hideous','appalling','atrocious','vile','abominable','frightful','sickening','nauseating','nauseous','stomach-churning','stomachturning','offputting','uninviting','unpalatable','unappetizing','unsavoury','distasteful','nasty','obnoxious',
'objectionable','odious','noxious','evil-smelling','foulsmelling','smelly','stinking','high','rank','rancid','fetid','malodorous','vomitous',
'ghastly','horrid','gruesome','godawful','gross','diabolical','beastly','putrid','yucky','icky','grotty','sick-making','gut-churning',
'whiffy','pongy','lousy','lousy','skanky','funky','on the nose','noisome','mephitic','disgustful','loathly','miasmic','miasmal','olid',
'dirty','filthy','mucky','grimy','grubby','stained','dirt-encrusted','muddy','muddied','unclean','unwashed','squalid','sordid','shabby',
'sleazy','soiled','sullied','scummy','rotten','defiled','decaying','putrefied','cruddy','manky','gungy','befouled','besmirched','begrimed',
'feculent']
```

```
foul_words = remove_punc.filter(lambda x : x in Foul_words)
foul_words.count()
```

```
foul_words.collect()
```

```
#### Counting violence words in the book

violence_words_list =['crash','terrorist','attack','blood','robbery','explosive','rage','rob','shooting','shoot','bomb','pistol','suicide','sword','savagery','cruelty','bloodthirstiness','viciousness','ferocity','callousness','heartlessness','mercilessness','pitilessness','remorselessness','sadism','inhumanity','heinousness','monstrousness','atrocity','vileness','kill','rape','sexual','abuse','molest','murder','probe','bestiality','brutishness','beastliness','blood','alcohol','retarded','psycho','vigilantism','exortion','racist','sever','force','violent','spoil','crazy','homicidal','psychotic','terrorism','chop','drugs','abduct']
```

```
violence_words = foul_words.filter(lambda x : x in violence_words_list)
violence_words.count()
```

```
violence_words.collect()
```

```
Explicit_words=['carnal','erotic','intimate','passionate','reproductive','mating','sensual','coitus','coition','copulation','fornication','carnal knowledge','sexual intercourse','make love','vulgar slangscrewing','fucking','shagging','fuck','screw','bang','lay','mate','seduce']
```

```
#### Counting of explicit words in the book

explicit_words = violence_words.filter(lambda x : x in Explicit_words)
explicit_words.count()
```

```
explicit_words.collect()
```

```
###	Counting the frequency of each foul words occuring

foulword_occur1 = foul_words.map(lambda e:(e,1))
foulword_occur2 = foulword_occur1.reduceByKey(lambda a,b:(a+b))
foulword_occur2.collect()
```

```
## Finding the percentage of foul words in the Book

foulword_occur = float(foul_words.count()/remove_punc.count())*100
foulword_occur
```
